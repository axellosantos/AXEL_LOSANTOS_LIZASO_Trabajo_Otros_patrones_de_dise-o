## 1. Memento

Es un patrón de diseño de comportamentos de objetos. Permite guardar estados de diferentes objetos y poder recuperarlos más adelante sin romper su encapsulación.

Esta se mantiene alamacendo para una un instante de tiempo en una clase independiente de aquella a la que pertenece el objeto, de forma que ese recuerdo permita que el objeto sea modificado y pueda volver a un estado anterior.

Los principales problemas que resulven son la restauración de objetos desde estados pasados y el hacer y deshacer determinadas operaciones, ya que para muchas aplicaciones esto es de vital importancia.

Se distinguen 3 clases importantes:

- Memento: Almacena el estado interno de un objeto originator. Puede almacenar todo o parte
- Originator: Crea un objeto Memento conteniendo un momento de su estado interno
- Caretaker: Solo almacena Mementos
  

Las consecuencias están relacionadas con la memoria. El uso frecuente de Mementos de gran tamaño, podría resultar costoso y perjudicar la velocidad del sistema.  
  
Las responsabilidades de originador al almacenar y mantener a salvo una o muchas copias de su estado podrían poner en peligro los datos.
  
  
      
	 

## 2. Adapter (Clases)


Es un patrón de diseño estructural de clases. Busca la manera de adaptar un objeto a otro de manera estandarizada.

Se presenta el problema cuando se quiere utilizar dos clases con interfaces distintas que son incompatibles. Tambien puede resultar que el objeto llamado no deba tener conocimiento de la otra clase de objetos.


Se distinguen 3 clases importantes:
- Target: Interfaz donde se incluyen los metodos objetivos
- Adapter: Es el que va a gacer como puente entre la clase que se va a adaptar y el Target
- Adaptado : Es la clase que se va a adaptar

Las consecuencias serian que ambas clases permanecen independientes unas de las otras. Tambien nos dan la posibilidad de un único adapter trabaje con muchos adapters

## 3. Flyweight

Es un patrón de diseño estructural de objetos que tiene la intencion de ahorrar espacio y utilizar el minimo de recursos necesarios.

Busca reducir o eliminar la redundancia cuanndo tenemos varios objetos con información idéntica, y para ello las referenciamos al mismo objeto.

Necesita dos estados: Estado intriseco y estado extrinseco

Se distinguen 3 clases importantes:
- Flyweight : almacenena el estado intrinseco el cual tiene que estar compartido y puede manipular el estado extrensico
- FlyweightFactory: crea y maneja lo objetos Flyweight y aseguran que estos se compartan.
- UnsharedConcreteFlyweight : Subclase de Flyweight que no tiene por que ser compartida. Permite que se comparta, pero no lo fuerza.
- ConcreteFlyweight: Implementa la interfaz Flyweight y almacena los estados compartidos.


Consecuencias como un aumento de costos en cuestion de tiempo en ejecución, un ahorro de almacenamiento que depende de la reducción de instancias y si el estado extrinseco es calculado o almacenado


## 4. Chain of Responsibility

Es un patrón de diseño de comportamiento de objetos. Permite establecer una cadena de objetos receptores a través de los cuales se pasa una petición formada por un objeto emisor. Cualquiera de estos receptores podrá responder a la petición en funcion de su criterio.

Se utiliza cuando se necesita enviar peticiones de un objeto que deben de ser atendidos por diferentes objetos. No se sabe cual objeto podrá resolver el problema

El problema era la necesidad de encontrar un sistema que pudiera servir a diferentes solicitudes de manera jerárquica.

Se distinguen 2 caracteristicas principales:

- Handler: Define la interfaz que trata las peticiones y contiene todo el arreglo de Handler hijos.
- HandlerConcreto: Trata las peticiones de las que es responsable. Si puede manejarla, lo hará. De no ser así pasaraá al siguiente sucesior.

Las consecuencias principales es que recude el acoplamiento, añade flexibilidad para asignar responsabilidades, pero no se garantiza la recepcion.


## 5. Prototype

Es un patrón de diseño de creación de objetos. Permite crear objetos prediseñados sin conocer detalles de como crearlo, sin necesidad de conocer su atributos o constructores. Los nuevos objetos que se crearán serán realmente clonados. 

El poblema viene cuando un sistemaposee objetos con datos repetitivos y queremos mejorar la velocidad. Es mucgo mas repido anivel del sistema clonarlos que crearlos.

Se distinguen 2 características principales:

- Prototype : Declara la interfaz del código que se clona.
- PrototypeConcreto : Implementa Prototype 


Como consecuencia un programa puede añadir y borrar objetos prototipo en tiempo de ejecución. Tambien se pueden especificar nuevos objetos prototipo variando los existentes y se desacopla la creación de las clases y la repetición de instanciación de objetos con parámetros repetitivos.
